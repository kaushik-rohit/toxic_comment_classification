```python
import pandas as pd
import numpy as np
```

```python
test = pd.read_csv('./test2.csv')
```

```python
test = test.drop(['Unnamed: 0'], axis=1)
test
```

```python
test['prediction'] = test.apply(lambda x: np.fromstring(x['prediction'][1:-1].strip(), dtype='float', sep=' '), axis=1)
```

```python
rows = []
for index, row in test.iterrows():
    id_ = row['id']
    preds = row['prediction']
    if len(preds) == 5:
        preds = [0, 0, 0, 0, 0, 0]
    rows += [[id_, preds[0], preds[1], preds[2], preds[3], preds[4], preds[5]]]
```

```python
df = pd.DataFrame(rows, columns=['id', 'toxic', 'severe_toxic', 'obscene', 'threat', 'insult', 'identity_hate'])
```

```python
df.to_csv('predictions.csv')
```
